```python
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(data['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(data['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(data['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(data['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(data['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(data['Secondary _Boys'])
```

```python
data['HrSecondary_Girls']=data['HrSecondary_Girls'].clip(lower=data['HrSecondary_Girls'].quantile(0.05), upper=data['HrSecondary_Girls'].quantile(0.95))
data['HrSecondary_Boys']=data['HrSecondary_Boys'].clip(lower=data['HrSecondary_Boys'].quantile(0.05), upper=data['HrSecondary_Boys'].quantile(0.95))
data['Upper Primary_Boys']=data['Upper Primary_Boys'].clip(lower=data['Upper Primary_Boys'].quantile(0.05), upper=data['Upper Primary_Boys'].quantile(0.95))
data['Upper Primary_Girls']=data['Upper Primary_Girls'].clip(lower=data['Upper Primary_Girls'].quantile(0.05), upper=data['Upper Primary_Girls'].quantile(0.95))
data['Secondary _Girls']=data['Secondary _Girls'].clip(lower=data['Secondary _Girls'].quantile(0.05), upper=data['Secondary _Girls'].quantile(0.95))
data['Secondary _Boys']=data['Secondary _Boys'].clip(lower=data['Secondary _Boys'].quantile(0.05), upper=data['Secondary _Boys'].quantile(0.95))
```

```python
# Outlier detection using boxplots 
plt.figure(figsize= (20,15))
plt.subplot(4,4,1)
sns.boxplot(data['HrSecondary_Girls'])

plt.subplot(4,4,2)
sns.boxplot(data['HrSecondary_Boys'])

plt.subplot(4,4,3)
sns.boxplot(data['Upper Primary_Boys'])

plt.subplot(4,4,4)
sns.boxplot(data['Upper Primary_Girls'])

plt.subplot(4,4,5)
sns.boxplot(data['Secondary _Girls'])

plt.subplot(4,4,6)
sns.boxplot(data['Secondary _Boys'])
```
